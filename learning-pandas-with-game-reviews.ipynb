# Count the number of games with Editor's choice "Y" using Boolean method 

```python
 NEditor = reviews[reviews.editors_choice =='Y'].shape[0]
print ("Number of games with Editor's choice Y are, ",NEditor)
```

# Count the number of games with Editor's choice "Y" on iPad

```python
 NEditorIpad = reviews[(reviews.editors_choice =='Y') & (reviews.platform =='iPad')].shape[0]
print ("Number of games with Editor's choice Y on Ipad are, ",NEditorIpad)
```

# Filter games on Xbox with score > 7 

```python
xbox_one_filter = reviews[(reviews['score']>7) & (reviews['platform'] =='Xbox One')]
xbox_one_filter.head()
```

# Map the Editor's choice with 1 (Y) and 0 (N)

```python
reviews['editors_choice'] = reviews.editors_choice.map({'Y':1,'N':0})
reviews.head()
```

# Percentage of total games Editors like

```python
# Count total number of games with 1
EdSum = reviews.editors_choice.sum()
# percentage of games with 1
print ("Percentage of games with Editor\'s choice are {0:.1f} %".format(EdSum/Nrows*100))
```
